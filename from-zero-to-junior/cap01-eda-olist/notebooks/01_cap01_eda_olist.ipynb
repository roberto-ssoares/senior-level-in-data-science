<font size="4" color='#CC403E'><strong>📌 2. O que é Análise Exploratória de Dados?</strong></font>

<font size="2" color='#66666'>

>- A Análise Exploratória de Dados é **o processo de investigar um conjunto de dados com o objetivo de compreender suas principais características** antes de aplicar técnicas mais avançadas de análise, modelagem ou visualização.

>- Ela responde perguntas como:
    - *Quantas linhas e colunas existem no dataset?*
    - *Quais são os tipos de variáveis?*
    - *Existem valores nulos?*
    - *Existem registros duplicados?*
    - *Existem outliers?*
    - *Como as variáveis estão distribuídas?*
    - *Há relação entre variáveis numéricas?*
    - *Existem padrões relevantes para o negócio?*

>- Em uma visão profissional, EDA não é apenas gerar gráficos.
    - ***EDA é transformar dados brutos em entendimento analítico.***

<font size="3" color='#CC403E'><strong>✔️ 01. Função auxiliar "find_project_root"</strong></font>

<font size="2" color='#66666'></font></div>

```python
def find_project_root(start_path: Path | None = None, marker_files: tuple[str, ...] = ("pyproject.toml", ".git")) -> Path:
    """
    Localiza a raiz do projeto procurando por arquivos/pastas marcadores,
    como pyproject.toml ou .git, subindo na árvore de diretórios.
    """
    current_path = Path.cwd() if start_path is None else Path(start_path).resolve()

    for path in [current_path, *current_path.parents]:
        if any((path / marker).exists() for marker in marker_files):
            return path

    raise FileNotFoundError(
        "Não foi possível localizar a raiz do projeto. "
        "Verifique se o notebook está dentro do repositório."
    )
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 08. Função de diagnóstico estrutural dos DataFrames</strong></font>

<font size="2" color='#66666'></font></div>

```python
def inspect_dataframe(df: pd.DataFrame, dataset_name: str) -> pd.DataFrame:
    """
    Cria um resumo estrutural de um DataFrame.

    Parâmetros
    ----------
    df : pd.DataFrame
        DataFrame a ser inspecionado.

    dataset_name : str
        Nome lógico do dataset.

    Retorno
    -------
    pd.DataFrame
        Tabela com informações estruturais por coluna.
    """
    
    total_rows = len(df)
    
    inspection = pd.DataFrame({
        "dataset": dataset_name,
        "column_name": df.columns,
        "dtype": df.dtypes.astype(str).values,
        "non_null_count": df.notna().sum().values,
        "null_count": df.isna().sum().values,
        "null_pct": (df.isna().mean() * 100).round(2).values,
        "unique_values": df.nunique(dropna=True).values,
        "unique_pct": ((df.nunique(dropna=True) / total_rows) * 100).round(2).values,
    })
    
    inspection["total_rows"] = total_rows
    inspection["duplicated_rows_in_dataset"] = df.duplicated().sum()
    
    return inspection
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 09. Inspeção estrutural de todos os datasets</strong></font>

<font size="2" color='#66666'></font></div>

```python
inspection_tables = []

for dataset_name, df in dfs.items():
    inspection_tables.append(
        inspect_dataframe(df=df, dataset_name=dataset_name)
    )

data_inspection = pd.concat(inspection_tables, ignore_index=True)

data_inspection.head(150)
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 10. Resumo executivo da inspeção estrutural</strong></font>

<font size="2" color='#66666'></font></div>

```python
inspection_summary = (
    data_inspection
    .groupby("dataset")
    .agg(
        total_rows=("total_rows", "max"),
        total_columns=("column_name", "count"),
        columns_with_nulls=("null_count", lambda x: (x > 0).sum()),
        total_null_values=("null_count", "sum"),
        max_null_pct=("null_pct", "max"),
        duplicated_rows=("duplicated_rows_in_dataset", "max"),
    )
    .reset_index()
    .sort_values(by="total_rows", ascending=False)
)

inspection_summary
```

```python
datetime_columns = {
    "orders": [
        "order_purchase_timestamp",
        "order_approved_at",
        "order_delivered_carrier_date",
        "order_delivered_customer_date",
        "order_estimated_delivery_date",
    ],
    "order_items": [
        "shipping_limit_date",
    ],
    "order_reviews": [
        "review_creation_date",
        "review_answer_timestamp",
    ],
}

datetime_columns
```

<font size="3" color='#CC403E'><strong>✔️ 15. Função para conversão controlada de datas</strong></font>

<font size="2" color='#66666'></font></div>

```python
def convert_datetime_columns(
    dataframes: dict[str, pd.DataFrame],
    datetime_map: dict[str, list[str]]
) -> dict[str, pd.DataFrame]:
    """
    Converte colunas temporais de múltiplos DataFrames para datetime.

    Parâmetros
    ----------
    dataframes : dict[str, pd.DataFrame]
        Dicionário contendo os DataFrames originais.

    datetime_map : dict[str, list[str]]
        Dicionário em que a chave é o nome lógico do dataset
        e o valor é a lista de colunas que devem ser convertidas.

    Retorno
    -------
    dict[str, pd.DataFrame]
        Novo dicionário com cópias dos DataFrames e colunas temporais convertidas.
    """

    typed_dataframes = {}

    for dataset_name, df in dataframes.items():
        df_typed = df.copy()

        if dataset_name in datetime_map:
            for column in datetime_map[dataset_name]:
                if column in df_typed.columns:
                    df_typed[column] = pd.to_datetime(
                        df_typed[column],
                        errors="coerce"
                    )

        typed_dataframes[dataset_name] = df_typed

    return typed_dataframes


dfs_typed = convert_datetime_columns(
    dataframes=dfs,
    datetime_map=datetime_columns
)

print("Conversão controlada de datas concluída.")
```

<font size="2" color='#66666'>

>- Criar variáveis derivadas a partir das datas dos pedidos para permitir análises temporais e logísticas.

>- **Ações realizadas**

<font size="3" color='#CC403E'><strong>✔️ 19. Criação de variáveis temporais em orders</strong></font>

<font size="2" color='#66666'></font></div>

```python
orders_enriched = orders.copy()

# Data base da compra
orders_enriched["purchase_date"] = orders_enriched["order_purchase_timestamp"].dt.date
orders_enriched["purchase_year"] = orders_enriched["order_purchase_timestamp"].dt.year
orders_enriched["purchase_month"] = orders_enriched["order_purchase_timestamp"].dt.month
orders_enriched["purchase_year_month"] = orders_enriched["order_purchase_timestamp"].dt.to_period("M").astype(str)
orders_enriched["purchase_day_of_week"] = orders_enriched["order_purchase_timestamp"].dt.day_name()
orders_enriched["purchase_hour"] = orders_enriched["order_purchase_timestamp"].dt.hour

# Tempo entre compra e aprovação
orders_enriched["approval_days"] = (
    orders_enriched["order_approved_at"] - orders_enriched["order_purchase_timestamp"]
).dt.total_seconds() / 86400

# Tempo real entre compra e entrega ao cliente
orders_enriched["delivery_days"] = (
    orders_enriched["order_delivered_customer_date"] - orders_enriched["order_purchase_timestamp"]
).dt.days

# Tempo estimado entre compra e data prevista de entrega
orders_enriched["estimated_delivery_days"] = (
    orders_enriched["order_estimated_delivery_date"] - orders_enriched["order_purchase_timestamp"]
).dt.days

# Diferença entre entrega real e entrega estimada
orders_enriched["delay_days"] = (
    orders_enriched["order_delivered_customer_date"] - orders_enriched["order_estimated_delivery_date"]
).dt.days

# Indicador de atraso
orders_enriched["is_delayed"] = orders_enriched["delay_days"] > 0

orders_enriched.head()
```

```python
orders_enriched["has_delivery_date"] = orders_enriched["order_delivered_customer_date"].notna()

orders_enriched["is_delayed"] = np.where(
    orders_enriched["has_delivery_date"],
    orders_enriched["delay_days"] > 0,
    np.nan
)

delay_quality_summary = pd.DataFrame(
    {
        "metric": [
            "total_orders",
            "orders_with_delivery_date",
            "orders_without_delivery_date",
            "delivered_on_time_or_early",
            "delivered_late",
            "delivered_late_pct_among_delivered",
        ],
        "value": [
            len(orders_enriched),
            orders_enriched["has_delivery_date"].sum(),
            (~orders_enriched["has_delivery_date"]).sum(),
            (orders_enriched["is_delayed"] == False).sum(),
            (orders_enriched["is_delayed"] == True).sum(),
            round(
                (orders_enriched["is_delayed"] == True).sum()
                / orders_enriched["has_delivery_date"].sum()
                * 100,
                2,
            ),
        ],
    }
)

delay_quality_summary
```

<font size="4" color='#CC403E'><strong>📌 26. Construção da Visão Comercial por Pedido</strong></font>

<font size="2" color='#66666'>

>- Construir uma visão comercial consolidada no nível do pedido, integrando informações de pedidos, itens e pagamentos.

>- **Ações realizadas**
    - *Avaliação da granularidade das tabelas `orders`, `order_items` e `order_payments`.*
    - *Agregação da tabela de itens para o nível do pedido.*
    - *Agregação da tabela de pagamentos para o nível do pedido.*
    - *Integração das tabelas agregadas com a tabela central de pedidos.*
    - *Validação da integridade da tabela analítica criada.*

>- **Justificativa técnica**
    - O dataset Olist possui tabelas com diferentes níveis de granularidade.
    - A tabela `orders` possui uma linha por pedido, enquanto `order_items` e `order_payments` podem possuir múltiplas linhas para o mesmo pedido.
    - Agregar essas tabelas antes do join evita duplicação indevida de valores financeiros.

>- **Resultados esperados**
    - Espera-se obter uma tabela analítica no nível do pedido, contendo informações de status, datas, quantidade de itens, valor dos produtos, valor do frete e valor total pago.

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 34. Análise de granularidade das tabelas comerciais</strong></font>

<font size="2" color='#66666'></font></div>

```python
granularity_summary = pd.DataFrame(
    {
        "dataset": [
            "orders_enriched",
            "order_items",
            "order_payments",
        ],
        "rows": [
            len(orders_enriched),
            len(order_items),
            len(order_payments),
        ],
        "unique_orders": [
            orders_enriched["order_id"].nunique(),
            order_items["order_id"].nunique(),
            order_payments["order_id"].nunique(),
        ],
    }
)

granularity_summary["rows_per_order_avg"] = (
    granularity_summary["rows"] / granularity_summary["unique_orders"]
).round(2)

granularity_summary
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 35. Agregação da tabela order_items no nível do pedido</strong></font>

<font size="2" color='#66666'></font></div>

```python
order_items_order_level = (
    order_items
    .groupby("order_id")
    .agg(
        items_count=("order_item_id", "count"),
        unique_products=("product_id", "nunique"),
        unique_sellers=("seller_id", "nunique"),
        items_gmv=("price", "sum"),
        freight_total=("freight_value", "sum"),
        avg_item_price=("price", "mean"),
    )
    .reset_index()
)

order_items_order_level["commercial_total"] = (
    order_items_order_level["items_gmv"] 
    + order_items_order_level["freight_total"]
)

order_items_order_level.head()
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 36. Agregação da tabela order_payments no nível do pedido</strong></font>

<font size="2" color='#66666'></font></div>

```python
def get_main_payment_type(series: pd.Series) -> str:
    """
    Retorna o tipo de pagamento mais frequente em um pedido.
    Em caso de empate, retorna o primeiro valor ordenado alfabeticamente.
    """
    valid_values = series.dropna()

    if valid_values.empty:
        return np.nan

    return valid_values.value_counts().sort_index().idxmax()


order_payments_order_level = (
    order_payments
    .groupby("order_id")
    .agg(
        payment_transactions=("payment_sequential", "count"),
        payment_value_total=("payment_value", "sum"),
        max_installments=("payment_installments", "max"),
        avg_installments=("payment_installments", "mean"),
        main_payment_type=("payment_type", get_main_payment_type),
        unique_payment_types=("payment_type", "nunique"),
    )
    .reset_index()
)

order_payments_order_level["avg_installments"] = (
    order_payments_order_level["avg_installments"].round(2)
)

order_payments_order_level.head()
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 37. Construção da tabela analítica comercial por pedido</strong></font>

<font size="2" color='#66666'></font></div>

```python
orders_commercial = (
    orders_enriched
    .merge(
        order_items_order_level,
        on="order_id",
        how="left"
    )
    .merge(
        order_payments_order_level,
        on="order_id",
        how="left"
    )
)

print(f"Linhas em orders_enriched:   {len(orders_enriched):,}")
print(f"Linhas em orders_commercial: {len(orders_commercial):,}")

print("\nPedidos únicos em orders_commercial:")
print(orders_commercial["order_id"].nunique())

print("\nPedidos duplicados em orders_commercial:")
print(orders_commercial["order_id"].duplicated().sum())

orders_commercial.head()
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 38. Validação da tabela analítica comercial</strong></font>

<font size="2" color='#66666'></font></div>

```python
commercial_validation = pd.DataFrame(
    {
        "metric": [
            "orders_rows_before_join",
            "orders_rows_after_join",
            "unique_orders_after_join",
            "duplicated_orders_after_join",
            "orders_without_items",
            "orders_without_payments",
        ],
        "value": [
            len(orders_enriched),
            len(orders_commercial),
            orders_commercial["order_id"].nunique(),
            orders_commercial["order_id"].duplicated().sum(),
            orders_commercial["items_count"].isna().sum(),
            orders_commercial["payment_value_total"].isna().sum(),
        ],
    }
)

commercial_validation
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 39. Resumo comercial geral</strong></font>

<font size="2" color='#66666'></font></div>

```python
commercial_summary = pd.DataFrame(
    {
        "metric": [
            "total_orders",
            "orders_with_items",
            "orders_with_payments",
            "total_items_gmv",
            "total_freight",
            "total_commercial_value",
            "total_payment_value",
            "avg_order_gmv",
            "median_order_gmv",
            "avg_freight_per_order",
            "freight_pct_over_gmv",
        ],
        "value": [
            orders_commercial["order_id"].nunique(),
            orders_commercial["items_count"].notna().sum(),
            orders_commercial["payment_value_total"].notna().sum(),
            round(orders_commercial["items_gmv"].sum(), 2),
            round(orders_commercial["freight_total"].sum(), 2),
            round(orders_commercial["commercial_total"].sum(), 2),
            round(orders_commercial["payment_value_total"].sum(), 2),
            round(orders_commercial["items_gmv"].mean(), 2),
            round(orders_commercial["items_gmv"].median(), 2),
            round(orders_commercial["freight_total"].mean(), 2),
            round(
                orders_commercial["freight_total"].sum()
                / orders_commercial["items_gmv"].sum()
                * 100,
                2,
            ),
        ],
    }
)

commercial_summary
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 40. Evolução mensal de pedidos, GMV e frete</strong></font>

<font size="2" color='#66666'></font></div>

```python
monthly_commercial = (
    orders_commercial
    .groupby("purchase_year_month")
    .agg(
        orders=("order_id", "nunique"),
        customers=("customer_id", "nunique"),
        items_gmv=("items_gmv", "sum"),
        freight_total=("freight_total", "sum"),
        payment_value_total=("payment_value_total", "sum"),
    )
    .reset_index()
    .sort_values("purchase_year_month")
)

monthly_commercial["avg_order_gmv"] = (
    monthly_commercial["items_gmv"] / monthly_commercial["orders"]
).round(2)

monthly_commercial["avg_freight_per_order"] = (
    monthly_commercial["freight_total"] / monthly_commercial["orders"]
).round(2)

monthly_commercial.head()
```

```python
payment_type_summary = (
    order_payments
    .groupby("payment_type")
    .agg(
        payment_transactions=("order_id", "count"),
        unique_orders=("order_id", "nunique"),
        payment_value_total=("payment_value", "sum"),
        avg_payment_value=("payment_value", "mean"),
        avg_installments=("payment_installments", "mean"),
    )
    .reset_index()
    .sort_values("payment_value_total", ascending=False)
)

payment_type_summary["payment_value_pct"] = (
    payment_type_summary["payment_value_total"]
    / payment_type_summary["payment_value_total"].sum()
    * 100
).round(2)

payment_type_summary["avg_payment_value"] = payment_type_summary["avg_payment_value"].round(2)
payment_type_summary["avg_installments"] = payment_type_summary["avg_installments"].round(2)

payment_type_summary
```

<font size="3" color='#CC403E'><strong>✔️ 45. Integração de itens, produtos e categorias</strong></font>

<font size="2" color='#66666'></font></div>

```python
items_products = (
    order_items
    .merge(
        products,
        on="product_id",
        how="left"
    )
    .merge(
        category_translation,
        on="product_category_name",
        how="left"
    )
)

print(f"Linhas em order_items:    {len(order_items):,}")
print(f"Linhas em items_products: {len(items_products):,}")

print("\nProdutos únicos:")
print(items_products["product_id"].nunique())

print("\nCategorias originais únicas:")
print(items_products["product_category_name"].nunique(dropna=True))

print("\nCategorias traduzidas únicas:")
print(items_products["product_category_name_english"].nunique(dropna=True))

items_products.head()
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 48. Resumo comercial por categoria de produto</strong></font>

<font size="2" color='#66666'></font></div>

```python
category_summary = (
    items_products
    .groupby("product_category_name_english")
    .agg(
        orders=("order_id", "nunique"),
        items_sold=("order_item_id", "count"),
        unique_products=("product_id", "nunique"),
        unique_sellers=("seller_id", "nunique"),
        items_gmv=("price", "sum"),
        freight_total=("freight_value", "sum"),
        avg_item_price=("price", "mean"),
        median_item_price=("price", "median"),
        avg_freight=("freight_value", "mean"),
    )
    .reset_index()
)

category_summary["commercial_total"] = (
    category_summary["items_gmv"] + category_summary["freight_total"]
)

category_summary["gmv_pct"] = (
    category_summary["items_gmv"] / category_summary["items_gmv"].sum() * 100
).round(2)

category_summary["avg_item_price"] = category_summary["avg_item_price"].round(2)
category_summary["median_item_price"] = category_summary["median_item_price"].round(2)
category_summary["avg_freight"] = category_summary["avg_freight"].round(2)

category_summary = category_summary.sort_values(
    by="items_gmv",
    ascending=False
)

category_summary.head(15)
```

<div style="background-color:#f3f2ee">
    
<font size="3" color='#CC403E'><strong>✔️ 71. Criação de grupos de avaliação</strong></font>

<font size="2" color='#66666'></font></div>

```python
def classify_review_score(score: float) -> str:
    """
    Classifica a nota média de avaliação em grupos interpretáveis.
    """
    if pd.isna(score):
        return "sem_avaliacao"
    elif score <= 2:
        return "negativa"
    elif score == 3:
        return "neutra"
    else:
        return "positiva"


orders_experience["review_group"] = orders_experience["avg_review_score"].apply(
    classify_review_score
)

review_group_order = ["negativa", "neutra", "positiva", "sem_avaliacao"]

review_group_summary = (
    orders_experience["review_group"]
    .value_counts(dropna=False)
    .reindex(review_group_order)
    .reset_index()
)

review_group_summary.columns = ["review_group", "orders"]

review_group_summary["orders_pct"] = (
    review_group_summary["orders"] / review_group_summary["orders"].sum() * 100
).round(2)

review_group_summary
```
